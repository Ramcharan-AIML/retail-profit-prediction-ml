# feature selection and dropping unwanted features

```python
df.columns.tolist()
```

```python
df['Sub-Category'].unique()
```

Grouping the Sub - Category column wiht some unique values with other_categories option

```python
sub_cat = df['Sub-Category'].value_counts()

rare_cat = sub_cat[sub_cat < 300].index.tolist()

df['Sub-Category'] = df['Sub-Category'].replace(rare_cat, "other_categories")

df['Sub-Category'].value_counts()
```

Grouping some unique values with others option

```python
least_states = df['State'].value_counts()
```

```python
rare_states = least_states[least_states < 100].index.tolist()
print(rare_states)

df['State'] = df['State'].replace(rare_states , "Others")

df['State'].value_counts()
print(len(df['State'].value_counts()))
```

```python
# Extracting the date and year from Order date column

df['Order Date'] = pd.to_datetime(df['Order Date'])

df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month

df.head(1)
```

```python
#  'Ship Mode', 'Segment', 'State', 'Region', 'Category', 'Sub-Category', 'Sales', 'Quantity', 'Discount', 'Profit'
```

```python
print(df.columns.tolist(), end=",")
```

```python
# Dropping the columns

df = df.drop(['Order Date','Row ID', 'Order ID', 'Ship Date','Country', 'Customer ID', 'Customer Name',  'City', 'Postal Code', 'Product ID', 'Product Name'],axis =1)

df
```

# Encoding the Categorical Features

```python
# Onehot Encoding 
df = pd.get_dummies(df , columns=['Ship Mode','Segment','State','Region','Category','Sub-Category'], drop_first=True)

```

```python
df.head(5)
```

# Training , Testing , Splitting the Data

```python
x = df.drop(columns= ['Profit'])
y = df['Profit']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test , y_train, y_test = train_test_split(x, y, test_size=0.2 , random_state=42)

```
