# Scaling the Data to maintain the Equal Cardinality

```python
# Standard Scaling 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scale_cols = ['Sales' , 'Quantity', 'Discount', 'Order Year', 'Order Month']

# Intializing the Scaler
scaler = StandardScaler()

# Fit and Transform the training data
# Fit can take Mean and Std_dvn and tranform can make scaling to the values
# we fit to learn the pattern and transform to apply it
X_train[scale_cols] = scaler.fit_transform(X_train[scale_cols])

# We can only transform the testing data here
# we use the rules learned from the train set to scale the test set
X_test[scale_cols] = scaler.transform(X_test[scale_cols])

X_train.head()
```

# Model Training and Predicting the Ouputs

```python
#importing the Linear regression Model

from sklearn.linear_model import LinearRegression

model_lr = LinearRegression()

# Train the model using our scaled training data
model_lr.fit(X_train , y_train)

# Ask the trained model to predict the answers(Profit) from with the unseen test data
y_pred_lr = model_lr.predict(X_test)

print("Our model Predictions :-", y_pred_lr[:5].round(2))
print("Actual Reality :-", y_test.values[:5].round(2))

```

# Metrics for the Model to Analyze the Overfitting and Underfitting

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , root_mean_squared_error, r2_score

# Ask the model to predict the training data like a practice test
y_pred_train = model_lr.predict(X_train)

# Calculating Metrics MAE , MSE , R2_score
def print_metrics(actual , predicted , dataset_name):
    mae = mean_absolute_error(actual , predicted )
    rmse = np.sqrt(mean_squared_error(actual , predicted ))
    r2 = r2_score(actual , predicted )

    print(f"{dataset_name} Performance")
    print(f'MAE (off by roughly): ${mae:.2f}')
    print(f'RMSE : ${rmse:.2f}')
    print(f'R-Squared Score: {r2:.2f}\n')

print_metrics(y_train, y_pred_train , "Training Data (Practice Test)")
print_metrics(y_test , y_pred_lr , "Testing Data (Final Exam)")

```

# Ridge and Lasso (Regularization) -- Controlling the Features

    Controlling the Beast (Regularization)
    Our Baseline Standard Linear Regression model failed. It has too much freedom to make crazy guesses based on 48 features.

    We need to put the puppy on a leash. This mathematical leash is called Regularization. It literally penalizes the model for having excessively large coefficients (weights).

    We will try two leashes:

    Ridge Regression (L2 Penalty): It shrinks all feature weights down closer to zero. It says to the model, "You can use all 48 features, but you aren't allowed to rely too heavily on any single one."
    Lasso Regression (L1 Penalty): It is much harsher. It also shrinks weights to zero, but it actually forces some weights to become exactly zero. It says to the model, "Half of these 48 features are garbage. I'm deleting them. You can only use the most important ones."

```python
from sklearn.linear_model import Ridge , Lasso

# Ridge Regression
# Alpha is how tight we pull the leash , Higher alpha = tighter leash
ridge_model = Ridge(alpha=10.0)
ridge_model.fit(X_train , y_train)

# Predict and Test
y_pred_ridge_test = ridge_model.predict(X_test)
r2_ridge_test = r2_score(y_test , y_pred_ridge_test)

# Lasso Regression
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train , y_train)

# Predict and score
y_pred_lasso_test = lasso_model.predict(X_test)
r2_lasso_test = r2_score(y_test , y_pred_lasso_test)


# Summary
print('-----------Final Exam Scores (Testing Data)--------------')
print(f'The baseline of R2 - Previously :- {-0.77}')

print(f'Ridge R2:- {r2_ridge_test}')
print(f'Lasso R2:- {r2_lasso_test}')
```

Inspecting the Lasso
Let's look inside the Lasso model you just trained to see exactly which features it kept and which ones it decided were completely useless.

```python
import pandas as pd

# 1. Grab the names of all 48 of our features
feature_names = X_train.columns

# 2. Grab the final weights (coefficients) the Lasso model assigned to them
lasso_weights = lasso_model.coef_

# 3. Create a nice DataFrame to view them side-by-side
lasso_df = pd.DataFrame({
    'Feature_Name': feature_names,
    'Weight': lasso_weights
})

# 4. Find the features Lasso "killed" (Weight exactly 0.0)
dead_features = lasso_df[lasso_df['Weight'] == 0.0]

# 5. Find the features Lasso kept (Weight NOT 0.0), sorted by importance
surviving_features = lasso_df[lasso_df['Weight'] != 0.0].copy()
# We look at the absolute value to see the overall impact (strong positive OR strong negative)
surviving_features['Absolute_Impact'] = abs(surviving_features['Weight'])
surviving_features = surviving_features.sort_values(by='Absolute_Impact', ascending=False)

print(f"Out of {len(feature_names)} original features, Lasso killed {len(dead_features)} of them!")
print("\n--- The Most Powerful Features Lasso Kept ---")
print(surviving_features[['Feature_Name', 'Weight']].head(10))
```

is lasso remove the 0.0 got features automaticaly when it is in model training and prediction time , is it automatic , it didn't take that features into the consideration while training and prediction time

```python
# Yes, exactly! It is 100% automatic.
# To understand how it's automatic, think about the simple math of how any Linear Regression model makes a prediction.
# The formula for predicting Profit looks like this: Predicted Profit = (Weight_1 * Sales) + (Weight_2 * Quantity) + (Weight_3 * Discount) ...
# During training, Lasso looks at a feature (let's pretend it's the state of Wyoming) and realizes, "This feature is just adding noise, it doesn't help me predict profit at all." Because of the "L1 Penalty" math built into Lasso, it takes the weight for Wyoming and forces it down to exactly 0.0.
# So, when predicting future data, the math looks like this: ... + (0.0 * Wyoming) + ...
# Because anything multiplied by zero is zero, that feature literally vanishes from the mathematical equation! It is permanently ignored for all future predictions.
# This is why Lasso is extremely popular in the medical field (e.g., if you have data on 10,000 human genes, Lasso will automatically delete the 9,900 genes that don't cause the disease, leaving you with only the important ones).
```
